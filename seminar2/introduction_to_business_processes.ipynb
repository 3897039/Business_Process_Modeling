- Опишите их главные отличия в бизнес-процессах
- Перечислите основные и вспомогательные бизнес-процессы

Возьмем для примера oz.by и OZON.
Главные отличия - реализация продукта. В OZ существуют магазины, преотавляющие часть продуктов с сайта. В OZON - только сайт с пунктами выдачи и почтометами.

Существуют основные и вспомогательные бизнес-процессы.

Основные процессы — это такие процессы, которые составляют суть бизнеса предприятия. 
Благодаря этим процессам компания получает прибыль. 
Также основные процессы формируют ценность для клиентов компании. 
Именно за них клиент компании готов платить.

В OZ и OZON это розничная или оптовая продажа.

Вспомогательные процессы еще называют бэк-офисом или поддерживающими процессами. 
Они также есть в любой компании и направлены на поддержание работы основных процессов.

Поддерживающие процессы убыточны, так как напрямую не приносят прибыль, 
но без них компания не сможет существовать.

Примерами поддерживающих процессов могут являться бухгалтерия, процессы по управлению персоналом,
обучение сотрудников, ведение задач по поддержке IT-инфраструктуры компании и так далее.
